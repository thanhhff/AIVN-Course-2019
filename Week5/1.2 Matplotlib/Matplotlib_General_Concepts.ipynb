## C 2 cch tip cn  v 1 Plot 
### Cch 1: S dng lnh Matplotlib c bn

Mt vi c php c bn:

- `plt.figure()`: to mt figure 


- `plt.plot()`: v  th vi gi tr trc x so vi trc y 


- `plt.xlabel()`: t tn cho trc x


- `plt.ylabel()`: t tn cho trc y


- `plt.title()`: t tn cho figure 


- `plt.grid()`: hin th cc ng li 


- `plt.legend()`: hin th ch thch cho  th 


- `plt.savefig()`: lu hnh v my 


- `plt.show()`: hin th hnh v 


- `plt.clf()`: xo hnh hin ti (hu ch khi c nhiu figure trong 1 chng trnh)

V d: v  th hm s: **`y = x^3`**

```python
import numpy as np
x = np.arange(-5, 5, 0.01)
y = x ** 3

# t tn 
plt.title('Hm y = x^3')
plt.xlabel('Trc x')
plt.ylabel('Trc y')

# Hin th li 
plt.grid()

# V plt.plot()
plt.plot(x, y, label='y = x^3')

# HIn th ch thch 
plt.legend()

# Hin th hnh v
plt.show()
```

Hn na, Matplotlib cn cho chng ta c th v nhiu hnh trn 1 Figure:
- Chng ta c th s dng phng thc `.subplot()` gm c 3 thng s c th:
    - `nrows`: s lng hng trn Figure
    
    - `ncols`: s lng ct trn Figure
      
    - `plot_number`: v tr ca plot trong Figure

```python
# plt.subplot(nrows, ncols, plot_number)

# nrows = 1, ncols = 2 (1 hng, 2 ct)
plt.subplot(1, 2, 1) # v  v tr 1
plt.plot(x, y, 'red')

plt.subplot(1, 2, 2) # v  v tr 2
plt.plot(x, y, 'green')

plt.show()
```

### Cch 2: Phng thc hng i tng

### Figure 
Nh  ni  trn, thnh phn chnh ca **figure** l cc **axes**. Mt **figure** c th cha mt hoc nhiu **axes**. Hay ni cch khc, **figure** ch l khung cha, chnh xc **axes** mi tht s l ni cc hnh c v ln.

Nh vy, s **khng c** nhng cch gi kiu nh:
    - fig.plot() 
    - fig.xlabel()
    - ...

```python
# Khi to figure trng
fig = plt.figure()

# Hin th figure hin ti 
plt.show()
```

By gi chng ta cn thm cc **axes** vo **figure**, bng phng thc: `.add_axes()` gm 4 i s: `bn tri, pha di, chiu rng, chiu cao.`

```python
fig = plt.figure()

# C th gi tt subplot(1,1,1) thnh subplot(111)
fig.add_subplot(111)

ax = fig.add_axes([0, 0.5, 0.5, 0.5], facecolor='#BECAFB')
ax.plot(x, y, 'red')

plt.show()
```

Thm tiu  cho cc axes bng cch:
   - `.set_xlabel()`: tiu  trc x
   
   - `.set_ylabel()`: tiu  trc y
   
   - `.set_title()`: tiu  axes

```python
fig = plt.figure()

# C th gi tt subplot(1,1,1) thnh subplot(111)
fig.add_subplot(111)

ax = fig.add_axes([0, 0.5, 0.5, 0.5], facecolor='#BECAFB')
ax.plot(x, y, 'red')

ax.set_xlabel('Trc x')
ax.set_ylabel('Trc y')
ax.set_title('Hnh th 1')

# To thm 1 axes trong cng 1 figure
ax2 = fig.add_axes([0.5, -0.1, 0.5, 0.5])

x = np.linspace(0, 2, 100)
ax2.plot(x, x, label='linear')
ax2.plot(x, x**2, label='quadratic')
ax2.plot(x, x**3, label='cubic')

ax2.set_xlabel('Trc x')
ax2.set_ylabel('Trc y')
ax2.set_title('Hnh th 2')
ax2.legend()

plt.show()
```

### Mt s thuc tnh khc ca Figure 
1. Chnh kch thc ca Figure bng phng thc `figsize` (n v inch).
2. `dpi`  phn gii ca hnh, gi tr mc nh l 100. `dpi` cng cao th cht lng nh cng cao.
3.1. Lu Figure bng `ig.savefig()`

```python
# Khi to nhanh Figure v Axes -> plt.subplots()
fig, ax = plt.subplots(figsize=(10,5))

x = np.linspace(0, 2, 100)
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')

ax.set_xlabel('Trc x')
ax.set_ylabel('Trc y')
ax.set_title('Hnh th 2')
ax.legend()

plt.show()

# Lu figure cng th mc vi code 
fig.savefig('my_figure.png')
```

3.2. Kim tra Figure  lu hay cha?

```python
import matplotlib.image as mpimg 

# Hin th hnh nh lu trong my 
plt.imshow(mpimg.imread('my_figure.png'))
```

## V  th kiu im 
Trong nhiu trng hp, ta mun hin th tng im thay v ng thng, vic chuyn ng thng sang im chng ta s dng thm cc `marker` trong `.plot()`.

Cc `marker` thng dng:


```python
t = np.arange(0., 10., 0.5)

plt.plot(t, t**2, 'X')
plt.show()
```

Cch v va im, va ng:
- Cch 1: ni cc im bng ng thng

Mt s kiu `linestyle`:


```python
x = [1, 2, 3, 5]
y = [4, 5, 6, 7]

plt.plot(x, y, linestyle='--', marker='o', color='b')

# Mt cch ngn hn 
plt.plot(x, y, '--bo')
plt.show()
```

- Cch 2: V mt vi im trn ng thng 

```python
x = np.linspace(-np.pi, np.pi, 30)
y = np.sin(x)
markers_on = [0, 19, 12, 5]
plt.plot(x, y, '-gD', markevery=markers_on)
plt.show()
```
